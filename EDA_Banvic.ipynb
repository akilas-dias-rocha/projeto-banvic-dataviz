# Analise Exploratoria - Banvic

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

PALETTE = sns.color_palette("Set2")
sns.set_palette(PALETTE)

repo_path = os.getcwd()


def explorar_tabela(df, nome):
    print(f"=== {nome} ===")
    print(f"Registros: {len(df)}")
    print(f"Duplicatas: {df.duplicated().sum()}")
    print(f"Valores nulos por coluna:")
    print(df.isnull().sum())
    print()
    df.info()
```

### Tabela "agencias"

```python
agencias = pd.read_csv(f"{repo_path}/banvic_dados/agencias.csv")
agencias.head()
```

```python
explorar_tabela(agencias, "agencias")
```

```python
agencias["data_abertura"] = pd.to_datetime(agencias["data_abertura"])
```

```python
agencias.head()
```

```python
quant_tipo_agencia = agencias["tipo_agencia"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=quant_tipo_agencia.index, y=quant_tipo_agencia.values, palette=PALETTE)
plt.title("Distribuicao de Tipos de Agencia", fontweight="bold")
plt.xlabel("Tipo de Agencia")
plt.ylabel("Total de Agencias")
plt.show()
```

```python
quant_cidades = agencias["cidade"].value_counts()
plt.figure(figsize=(10, 4))
sns.barplot(x=quant_cidades.index, y=quant_cidades.values, palette=PALETTE)
plt.title("Distribuicao de Agencias por Cidade", fontweight="bold")
plt.xlabel("Cidades")
plt.ylabel("Total de Agencias")
plt.show()
```

**Achados - agencias:** A tabela possui 10 registros sem duplicatas. Ha dois tipos de agencia: Fisica (majoritaria) e Digital. Todas as agencias estao localizadas em cidades do estado de Sao Paulo, com a cidade de Sao Paulo concentando a maior parte das agencias.

```python
clientes = pd.read_csv(f"{repo_path}/banvic_dados/clientes.csv")
clientes.head()
```

### Tabela "contas"

```python
contas = pd.read_csv(f"{repo_path}/banvic_dados/contas.csv")
contas.head()
```

```python
explorar_tabela(contas, "contas")
```

```python
contas["data_abertura"] = pd.to_datetime(contas["data_abertura"], format="mixed")
contas["data_ultimo_lancamento"] = pd.to_datetime(contas["data_ultimo_lancamento"], format="mixed")
```

```python
contas.describe()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=contas, x="saldo_total", bins=10, color=PALETTE[0])
plt.title("Distribuicao de Saldo Total", fontweight="bold")
plt.xlabel("Saldo Total (R$)")
plt.ylabel("Numero de Contas")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=contas, x="saldo_total", color=PALETTE[1])
plt.title("Distribuicao de Saldo Total", fontweight="bold")
plt.xlabel("Saldo Total (R$)")
plt.show()
```

**Achados - contas:** A tabela possui aproximadamente 999 contas sem duplicatas. A media do saldo total e superior a mediana, indicando uma distribuicao assimetrica a direita com presenca de outliers (contas com saldos muito elevados). O saldo disponivel e proximo ao saldo total na maioria dos casos.

### Tabela "transacoes"

```python
transacoes = pd.read_csv(f"{repo_path}/banvic_dados/transacoes.csv")
transacoes.head()
```

```python
explorar_tabela(transacoes, "transacoes")
```

```python
transacoes["data_transacao"] = pd.to_datetime(transacoes["data_transacao"], format="mixed")
```

```python
transacoes.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(y=transacoes["nome_transacao"], order=transacoes["nome_transacao"].value_counts().index, palette=PALETTE)
plt.title("Contagem de Transacoes por Tipo", fontweight="bold")
plt.xlabel("Numero de Transacoes")
plt.ylabel("Tipo de Transacao")
plt.show()
```

**Achados - transacoes:** A tabela registra transacoes bancarias nos tipos Credito, Debito, Transferencia e Deposito. O tipo Credito e o mais frequente. A data da transacao foi convertida para datetime.

### Exportando tabelas tratadas

```python
output_dir = f"{repo_path}/banvic_limpo"
os.makedirs(output_dir, exist_ok=True)

dataframes = {
    "agencias": agencias,
    "clientes": clientes,
    "colaborador_agencia": colaborador_agencia,
    "colaboradores": colaboradores,
    "contas": contas,
    "propostas_credito": propostas_credito,
    "transacoes": transacoes,
}

for nome, df in dataframes.items():
    df.to_csv(f"{output_dir}/{nome}.csv", index=False)

print("Processo finalizado")
```
