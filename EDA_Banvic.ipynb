# Análise Exploratória - Banvic

```python
# Configuracao do ambiente de análise
# Projeto: Análise Exploratória de Dados - BanVic
# Período dos dados: 01/01/2010 ate 15/01/2023
# Objetivo: Identificar insights e indicadores para o dashboard
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

PALETTE = sns.color_palette("Set2")
sns.set_palette(PALETTE)

repo_path = os.getcwd()

# Função para overview da tabela
def explorar_tabela(df, nome):
    print(f"=== {nome} ===")
    print(f"Registros: {len(df)}")
    print(f"Duplicatas: {df.duplicated().sum()}")
    print(f"Valores nulos por coluna:")
    print(df.isnull().sum())
    print()
    df.info()
```

### Tabela "agencias"

```python
# Importando tabela de Agências

agencias = pd.read_csv(f"{repo_path}/banvic_dados/agencias.csv")
agencias.head()
```

```python
explorar_tabela(agencias, "agencias")
```

```python
agencias["data_abertura"] = pd.to_datetime(agencias["data_abertura"])
```

```python
# Analisando a distribuição de agências por tipo

quant_tipo_agencia = agencias["tipo_agencia"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=quant_tipo_agencia.index, y=quant_tipo_agencia.values, palette=PALETTE)
plt.title("Distribuição de Tipos de Agência", fontweight="bold")
plt.xlabel("Tipo de Agência")
plt.ylabel("Total de Agências")
plt.show()
```

```python
# Analisando a distribuição de agências por cidade

quant_cidades = agencias["cidade"].value_counts()
plt.figure(figsize=(10, 4))
sns.barplot(x=quant_cidades.index, y=quant_cidades.values, palette=PALETTE)
plt.title("Distribuição de Agências por Cidade", fontweight="bold")
plt.xlabel("Cidades")
plt.ylabel("Total de Agências")
plt.show()
```

**Achados - agencias:** 
- Total de 10 agências, sendo 9 físicas e 1 digital
- Distribuição por cidade: São Paulo (4), Campinas (1), Osasco (1), Porto Alegre (1), Rio de Janeiro (1), Florianópolis (1), Recife (1)
- A tabela possui 10 registros sem duplicatas e não possui campos nulos

### Tabela "clientes"

```python
clientes = pd.read_csv(f"{repo_path}/banvic_dados/clientes.csv")
clientes.head()
```

```python
explorar_tabela(clientes, "clientes")
```

```python
# Convetendo as colunas de data para o tipo 'datetime'

clientes["data_inclusao"] = pd.to_datetime(clientes["data_inclusao"])
clientes["data_nascimento"] = pd.to_datetime(clientes["data_nascimento"])
```

```python
# Analisando a data de inclusao dos clientes

ultimo_cliente = clientes["data_inclusao"].max()
primeiro_cliente = clientes["data_inclusao"].min()
print(f"Data de inclusão do cliente mais recente: {ultimo_cliente}")
print(f"Data de inclusão do cliente mais antigo: {primeiro_cliente}")
```

```python
# Analisando a distribuição de tipos de clientes

quant_clientes = clientes["tipo_cliente"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=quant_clientes.index, y=quant_clientes.values, palette=PALETTE)
plt.title("Distribuição de Tipos de Clientes", fontweight="bold")
plt.xlabel("Tipo de Cliente")
plt.ylabel("Total de Clientes")
plt.show()
```

```python
# Analisando a idade dos clientes

clientes["idade"] = (pd.Timestamp("now") - clientes["data_nascimento"]).dt.days // 365
print(f"Idade do cliente mais novo: {clientes['idade'].min()}")
print(f"Idade do cliente mais velho: {clientes['idade'].max()}")
```

```python
# Gráfico histograma de idade dos clientes

faixas_idade = [18, 30, 42, 54, 66, 72, 84]
labels_faixas = ["18-30", "30-42", "42-54", "54-66", "66-72", "72-84"]

clientes["faixa_idade"] = pd.cut(clientes["idade"], bins=faixas_idade, labels=labels_faixas, right=False)

plt.figure(figsize=(10, 4))
sns.countplot(x=clientes["faixa_idade"], palette=PALETTE, order=labels_faixas)
plt.title("Distribuição de Idade dos Clientes", fontweight="bold")
plt.xlabel("Faixa de Idade")
plt.ylabel("Número de Clientes")
plt.show()
```

**Achados - clientes:** 
- 998 registros, todos de Pessoa Fisica (PF)
- Clientes realizam transações diárias através de diferentes formas de pagamento
- Perfis diversos com diferentes demandas de serviços financeiros
- Colunas de data convertidas: data_inclusao e data_nascimento

### Tabela "contas"

```python
contas.describe()
```

**Achados - contas:**

- Média de saldo: R$11.668, Desvio padrao: R$41.872 (alta dispersão)
- Saldo máximo: R$419.923 (presença de outliers significativos)
- A média do saldo total e superior a mediana, indicando distribuição assimétrica a direita
