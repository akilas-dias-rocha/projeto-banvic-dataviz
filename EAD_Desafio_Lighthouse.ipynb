# Análise Exploratória - Banvic

### Tabela "agencias"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

agencias = pd.read_csv("E:/Documentos/Desafio Lighthouse/banvic_data/agencias.csv")
agencias.head()
```

```python
agencias.info()
```

```python
agencias.duplicated().sum()
```

```python
agencias["data_abertura"] = pd.to_datetime(agencias["data_abertura"])
```

```python
agencias.head()
```

```python
quant_tipo_agencia = agencias["tipo_agencia"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=quant_tipo_agencia.index, y=quant_tipo_agencia.values)
plt.title("Distribuição de Tipos de Agência", fontweight="bold")
plt.xlabel("Tipo de Agência")
plt.ylabel("Total de Agências")
plt.show()
```

```python
quant_cidades = agencias["cidade"].value_counts()
plt.figure(figsize=(10, 4))
sns.barplot(x=quant_cidades.index, y=quant_cidades.values)
plt.title("Distribuição de Agências por Cidade", fontweight="bold")
plt.xlabel("Cidades")
plt.ylabel("Total de Agências")
plt.show()
```

### Tabela "contas"
